```python
# ==============================
# SILVER LAYER - EXCHANGE CLEANING
# ==============================

from pyspark.sql.functions import col

# ==============================
# STEP 1: READ FROM BRONZE (DATE PARTITION)
# ==============================

df_exchange = spark.read.parquet("Files/Bronze/processed/2026-05-20/exchange_rate")

# ==============================
# STEP 2: CLEAN DATA
# ==============================

valid_exchange = df_exchange.filter(
    col("exchange_rate").isNotNull()
)

invalid_exchange = df_exchange.subtract(valid_exchange)

# ==============================
# STEP 3: WRITE TO SILVER
# ==============================

valid_exchange.write.mode("overwrite").saveAsTable("silver_exchange_rate")
invalid_exchange.write.mode("overwrite").saveAsTable("silver_exchange_rate_quarantine")

# ==============================
# STEP 4: VALIDATION
# ==============================

print("Valid exchange:", valid_exchange.count())
print("Invalid exchange:", invalid_exchange.count())

# ==============================
# EXIT PIPELINE
# ==============================

mssparkutils.notebook.exit("Success")
```
