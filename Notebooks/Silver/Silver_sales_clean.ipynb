```python
# ==============================
# SILVER LAYER - SALES CLEANING
# ==============================

from pyspark.sql.functions import col

# ==============================
# STEP 1: READ FROM BRONZE (DATE PARTITION)
# ==============================

df_sales = spark.read.parquet("Files/Bronze/processed/2026-05-20/sales")

# ==============================
# STEP 2: CLEAN DATA
# ==============================

valid_sales = df_sales.filter(
    (col("order_status") != "Failed") &
    (col("feedback_score").between(1, 5))
)

invalid_sales = df_sales.subtract(valid_sales)

# ==============================
# STEP 3: WRITE TO SILVER
# ==============================

valid_sales.write.mode("overwrite").saveAsTable("silver_sales")
invalid_sales.write.mode("overwrite").saveAsTable("silver_sales_quarantine")

# ==============================
# STEP 4: VALIDATION
# ==============================

print("Valid sales:", valid_sales.count())
print("Invalid sales:", invalid_sales.count())

# ==============================
# EXIT PIPELINE
# ==============================

mssparkutils.notebook.exit("Success")
```
