```python
from pyspark.sql.functions import current_timestamp

# ==============================
# READ FROM BRONZE (DATE PARTITION)
# ==============================

df_sales = spark.read.csv("Files/Bronze/2026-05-20/sales.csv", header=True, inferSchema=True)
df_exchange = spark.read.csv("Files/Bronze/2026-05-20/exchange_rate.csv", header=True, inferSchema=True)

# ==============================
# ADD METADATA
# ==============================

df_sales = df_sales.withColumn("ingestion_time", current_timestamp())
df_exchange = df_exchange.withColumn("ingestion_time", current_timestamp())

# ==============================
# SAVE AS PARQUET
# ==============================

df_sales.write.mode("overwrite").parquet("Files/Bronze/processed/2026-05-20/sales")
df_exchange.write.mode("overwrite").parquet("Files/Bronze/processed/2026-05-20/exchange_rate")

# ==============================
# VALIDATION
# ==============================

print("Sales count:", df_sales.count())
print("Exchange count:", df_exchange.count())

print("Bronze ingestion completed")

# ==============================
# EXIT PIPELINE
# ==============================

mssparkutils.notebook.exit("Success")
```
